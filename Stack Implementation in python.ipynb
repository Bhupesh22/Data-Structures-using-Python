```python
class Stack:
    def __init__(self):
        self.items=[]
    
    def isempty(self):
        if self.items==[]:
            return -1
    
    def push(self,data):
        maximum=4
        if len(self.items)==maximum:
            print('Stack Overflow ...')
            return
        else:
            return self.items.append(data)
    
    def pop1(self):
        if self.isempty()==-1:
            print('Stack Underflow...')
            return
        else:
            return self.items.pop()
    
    def stacktop(self):
        return self.items[-1]
    
    def display(self):
        print(self.items[::-1])

        
s=Stack()
print("Enter Stack Operations:\n1:Push\n2:Pop\n3:Stacktop\n4:Print Stack\n5:quit\n-------------------\n")
while(1):
    x=int(input("Enter choices:"))
    if x==1:
        print("\nEnter element to push:")
        ele=int(input())
        s.push(ele)
        s.display()
    if x==2:
        y=s.pop1()
        print("\nPopped element:",y)
        s.display()
    if x==3:
        y=s.stacktop()
        print("Stack top is : ",y)
    if x==4:
        print('Stack display')
        s.display()
    if x==5:
        print('No operations ! Quit .')
        break

```
