```python
class Queue:
    def __init__(self):
        self.item=[]
        self.f=0
        self.r=-1
    
    def insert(self,data):
        MAX=5
        if self.r==MAX-1:
            print("Queue overflow...")
            return
        self.r=self.r+1
        self.item.append(data)
    
    def isempty(self):
        return self.item==[]
    
    def delete(self):
        if self.isempty():
            print("Queue Underflow")
            return
        z=self.item[self.f]
        self.item.remove(z)
        return z
    
    def front(self):
        if self.isempty():
            print("Queue Empty")
            return
        return self.item[self.f]
    
    def rear(self):
        if self.isempty():
            print("Queue Empty")
            return
        return self.item[-1]
    
    def display(self):
        if self.isempty():
            print("Queue Empty")
            return
        print(self.item[:])

    
q=Queue()
print("Enter Queue Operations:\n1:Insert\n2:Delete\n3:Queue Front\n4:Queue Rear\n5:Display\n6:Quit\n-------------------\n")
while(1):
    x=int(input("Enter choices:"))
    if x==1:
        print("\nEnter element to insert:")
        ele=int(input())
        q.insert(ele)
        q.display()
    if x==2:
        y=q.delete()
        print("\nDeleted element:",y)
        q.display()
    if x==3:
        y=q.front()
        print("\nQueue front is : ",y)
    if x==4:
        y=q.rear()
        print("\nQueue rear is : ",y)
    if x==5:
        print('\nQueue display',end=' ')
        q.display()
    if x==6:
        print('No operations ! Quit .')
        break

```
