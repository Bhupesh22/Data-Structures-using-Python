```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class LinkedList:
    def __init__(self):
        self.start=None
    
    def insertbeg(self,data):
        newnode=Node(data)
        newnode.next=self.start
        self.start=newnode
    
    def insertend(self,data):
        newnode=Node(data)
        if self.start is None:
            self.start=newnode
            return
        x=self.start
        while(x.next is not None):
            x=x.next
        x.next=newnode
    
    def insertpos(self,pos,data):
        newnode=Node(data)
        x=self.start
        for i in range(0,pos-2):
            x=x.next
        newnode.next=x.next
        x.next=newnode
        
    def insertbeforeitem(self,item,data):
        newnode=Node(data)
        if self.start is None:
            print("Linked List Empty\n")
            return
        if self.start.data==item:
            newnode.next=self.start
            self.start=newnode
            return
        
        x=self.start
        while(x.next is not None):
            if x.next.data==item:
                break
            x=x.next
        if x.next is None:
            print("Item not in the list\n")
        else:
            newnode=Node(data)
            newnode.next=x.next
            x.next=newnode
    
    def insertafteritem(self,item,data):
        newnode=Node(data)
        x=self.start
        while(x.next is not None):
            if x.data==item:
                break
            x=x.next
        if x.next is None:
            print("Item not in list\n")
            return
        newnode.next=x.next
        x.next=newnode
        
    def deletebeg(self):
        if self.start is None:
            print("Linked List Empty\n")
            return
        self.start=self.start.next
        
    def deleteend(self):
        if self.start is None:
            print("Linked List Empty\n")
            return
        x=self.start
        while(x.next.next is not None):
            x=x.next
        x.next=None
        
    def deletebyvalue(self,item):
        if self.start==None:
            print("LinkedList empty\n")
            return 
        if self.start.data==item:
            self.start=self.start.next
        x=self.start
        while(x.next is not None):
            if(x.next.data==item):
                break
            x=x.next
        if x.next is None:
            print("Item not found\n")
            return
        x.next=x.next.next
        
    def count(self):
        countlist=0
        x=self.start
        while(x is not None):
            countlist=countlist+1
            x=x.next
        return countlist
    
    def reverse(self):
        previous=None
        current=self.start
        while(current is not None):
            nextx=current.next
            current.next=previous
            previous=current
            current=nextx
        self.start=previous
    
    def display(self):
        if self.start is None:
            print("Linked List empty\n")
            return
        x=self.start
        while(x is not None):
            print(x.data,end=' ')
            x=x.next
        print("\n")
            
l=LinkedList()

#Insert Operations
print("----------Insert Operations--------\n")
l.deletebeg()   #just to check condition 
l.insertbeg(20)
l.display()
l.insertbeg(30)
l.display()
l.insertbeg(90)
l.display()
l.insertend(50)
l.display()
l.insertend(60)
l.display()
l.insertend(67)
l.display()
l.insertpos(3,56)
l.display()
l.insertbeforeitem(50,78)
l.display()
l.insertafteritem(78,80)
l.display()
print("**To check for element not found during inserting before item \n")
l.insertbeforeitem(99,2)
print("**To check for element not found during inserting after item \n")
l.insertafteritem(99,2)

#delete operations
print("----------Delete Operations-----------\n")
l.deletebeg()
l.display()
l.deleteend()
l.display()
l.deletebyvalue(78)
l.display()
print("**To check for item which is not in linked list deletebyvalue(99)\n")
l.deletebyvalue(99)

#Other opertions
print("--------------Other Operations---------\n")
c=l.count()
print("Total elements in the linked list are :",c)
print("\nReversing linked list : ",end=' ')
l.reverse()
l.display()
```
