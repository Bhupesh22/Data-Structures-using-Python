```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

class BinarySearchTree:
    def __init__(self):
        self.root=None
    
    def insert(self,ele):
        p=Node(ele)
        if self.root is None:
            self.root=p
            return
        q=self.root
        while(q is not None):
            if ele<=q.data:
                if q.left is None:    
                    q.left=p
                    return
                else:
                    q=q.left
            else:
                if q.right is None:
                    q.right=p
                    return
                else:
                    q=q.right
    
    def inorder(self,x):
        if x is not None:
            self.inorder(x.left)
            print(x.data,end='-')
            self.inorder(x.right)
    
    def preorder(self,x):
        if x is not None:
            print(x.data,end='-')
            self.preorder(x.left)
            self.preorder(x.right)
    
    def postorder(self,x):
        if x is not None:
            self.postorder(x.left)
            self.postorder(x.right)
            print(x.data,end='-')
    
    def bstmax(self,x):
        if x.right==None:
            return x.data
        else:
            return self.bstmax(x.right)
    
    def bstmin(self,x):
        if x.left==None:
            return x.data
        else:
            return self.bstmin(x.left)
        
    def countnodes(self,x):
        self.count=1
        if x.left is not None:
            self.count=self.count+self.countnodes(x.left)
        if x.right is not None:
            self.count=self.count+self.countnodes(x.right)
        return self.count
    
    def countleaf(self,x):
        self.countl=0
        if x.left==None and x.right==None:
            self.countl=1
        else:
            if x.left is not None:
                self.countl=self.countl+self.countleaf(x.left)
            if x.right is not None:
                self.countl=self.countl+self.countleaf(x.right)
        return self.countl
    
    def countinternalnode(self,x):
        if x==None or (x.left ==None and x.right==None):
            return 0
        return 1+self.countinternalnode(x.left)+self.countinternalnode(x.right)
    
    def maxheight(self,x):
        if x is None:
            return 0
        lheight=self.maxheight(x.left)
        rheight=self.maxheight(x.right)
        if lheight>rheight:
            return lheight+1
        else:
            return rheight+1
    
    
    def search(self,ele):
        q=self.root
        while(q is not None):
            if ele==q.data:
                return 1
            if ele<=q.data:
                q=q.left
            else:
                q=q=q.right
        return 0
    
    def printpath(self,x,path):
        if x is None:
            return
        path.append(x.data)
        self.printpath(x.left,path)
        if x.left==None and x.right==None:
            print(path)
        self.printpath(x.right,path)
        path.pop()
    
    def printpathsum(self,x,pathsum,ele):
        if x is None:
            return
        pathsum.append(x.data)
        self.printpathsum(x.left,pathsum,ele)
        if x.left==None and x.right==None:
            if sum(pathsum)==ele:
                print(pathsum)
        self.printpathsum(x.right,pathsum,ele)
        pathsum.pop()
    
    def father(self,x,ele):
        if x is None:
            return False
        if x.data==ele:
            return True
        if ele<=x.data:
            if(self.father(x.left,ele)):
                print(x.data)
        else:
            if(self.father(x.right,ele)):
                print(x.data)

    
b=BinarySearchTree()
b.insert(57)
b.insert(29)
b.insert(10)
b.insert(47)
b.insert(37)
b.insert(42)
b.insert(53)
b.insert(80)
b.insert(65)
b.insert(75)
b.insert(97)
b.insert(70)
b.insert(76)
print('--------------------------------------')
print('Inorder Traversal :')
b.inorder(b.root)
print('\n------------------------------------')
print('\nPreorder Traversal :')
b.preorder(b.root)
print('\n------------------------------------')
print('\nPostorder Traversal :')
b.postorder(b.root)
print('\n------------------------------------')
m=b.bstmax(b.root)
print('Maximum element is BST: ',m)
print('\n------------------------------------')
mn=b.bstmin(b.root)
print('Minimum element is BST: ',mn)
print('\n-------------------------------------')
cnt=b.countnodes(b.root)
print('Total number of nodes in BST :',cnt)
print('\n-------------------------------------')
cntl=b.countleaf(b.root)
print('Total Leaf Node in BST :',cntl)
print('\n-------------------------------------')
cntint=b.countinternalnode(b.root)
print('Total Non-Leaf(Internal) Node in BST :',cntint)
print('\n-------------------------------------')
h=b.maxheight(b.root)
print('Height of tree is ',h)
print('\n-------------------------------------')

ele=int(input('Enter element to search:'))
res=b.search(ele)
if res==1:
    print('Element present')
else:
    print('Element not present')
print('\n------------------------------------')
stack=[]
print('All paths in BST are :\n')
b.printpath(b.root,stack)
print('\n------------------------------------')
pathsum=[]
k=int(input('Enter the element to find the path whose sum is equal to = :'))
b.printpathsum(b.root,pathsum,k)
print('\n------------------------------------')
ele=int(input('Enter the element to find the father of :'))
b.father(b.root,ele)
```
